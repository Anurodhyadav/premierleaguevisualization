```python
#%matplotlib inline
%matplotlib nbagg
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv(r'F:\Downloads\premier-league\results.csv')
df.head()
```

```python
df.tail()
```

```python
pf=df.loc[(df['home_team']=='Liverpool') | (df['away_team']=='Liverpool'),:]
pf.head(38)
```

```python
for index,row in pf.iterrows():
    if (row['home_team']=='Liverpool'):
        pf.loc[index,'Goal_scored']=row['home_goals']
        pf.loc[index,['Goal_conceded']]=row['away_goals']
        #pf.at[index,'Goal_scored']=row['home_goals'] #Both loc code and at code does the same job here,we can use any
        #pf.at[index,'Goal_conceded']=row['away_goals']
    elif (row['away_team']=='Liverpool'):
        pf.at[index,'Goal_scored']=row['away_goals']
        pf.at[index,'Goal_conceded']=row['home_goals']
        
        
        
    
    

    
```

```python
pf.head()
```

```python
seasons=list(set(pf['season']))
```

```python
seasons
```

```python
seasonsort=seasons.sort()
```

```python
seasons
```

```python
sum=np.zeros(12)
for i,j in enumerate(seasons):
    lf=pf.loc[pf['season']==j,['Goal_scored']]
    sum[i]=lf['Goal_scored'].sum()
    
    
    
```

```python
sum
```

```python
dif=np.zeros(12)
for i,j in enumerate(seasons):
    lf=pf.loc[pf['season']==j,['Goal_conceded']]
    dif[i]=lf['Goal_conceded'].sum()
    
```

```python
dif
```

```python
plt.figure(figsize=(42,20))
ax=plt.gca()
x=np.arange(12)
width=0.4
rects1=ax.bar(x-width/2,sum,width,label='Goals-scored')
rects2=ax.bar(x+width/2,dif,width,label='Conceded-goals')
ax.legend()
ax.set_xlabel('Seasons')
ax.set_ylabel('Goals')
ax.set_xticks(x)
ax.set_title('Goals Scored and Concede by Liverpool  in the Premier League from 2006-07 to 2017-18')
ax.set_xticklabels(seasons)

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

plt.show()
```

```python
for index,row in pf.iterrows():
    if (row['Goal_scored']==row['Goal_conceded']):
        pf.loc[index,'Result']='D'
    elif(row['Goal_scored']>row['Goal_conceded']):
        pf.loc[index,'Result']='W'
    else:
        pf.loc[index,'Result']='L'
        
pf.head()
```

```python
result=pf['Result'].value_counts()
```

```python
result=result.tolist()
```

```python
result
```

```python
colors = ["#E13F29", "#D69A80", "#D63B59"]
fig,ax=plt.subplots()
ax.pie(result,labels=['Win','Draw','Loss'],shadow=True,colors=colors,startangle=90,explode=(0,0,0.05),autopct='%1.1f%%',)
ax.axis('equal')

# View the plot
plt.tight_layout()
plt.title('Liverpool record in premier league from 2006-07 to 2017-18',y=0)
plt.show()



```
